# Let's perform PCA and plot the result to see the distribution of subtypes

```python
mirna_train_final_subtype = mirna_train_final_subtype.rename(columns={"Ciriello_subtype": "subtype"})
mirna_test_final_subtype = mirna_test_final_subtype.rename(columns={"expert_PAM50_subtypes": "subtype"})
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)

mirna_all = pd.concat([mirna_train_final_subtype, mirna_test_final_subtype], sort=True, ignore_index=True, axis=0)
mirna_all_target = mirna_all["subtype"]
mirna_all_num = mirna_all.drop(["subtype"], axis=1)
mirna_all_comp = pca.fit_transform(mirna_all_num)
```

```python
mirna_all_comp_df = pd.DataFrame(mirna_all_comp, columns = ['principal component 1', 'principal component 2'])
print(mirna_all_comp_df.shape)
finalDf = pd.concat([mirna_all_comp_df, mirna_all_target], axis="columns")
```

```python
fig = plt.figure(figsize = (8,8))
ax = fig.add_subplot(1,1,1) 
ax.set_xlabel('Principal Component 1', fontsize = 15)
ax.set_ylabel('Principal Component 2', fontsize = 15)
ax.set_title('2 component PCA', fontsize = 20)
targets = ['Normal', 'Basal', 'LumA', 'LumB', 'Her2']
colors = ['r', 'orange', 'k', 'g', 'c']
for target, color in zip(targets,colors):
    indicesToKeep = finalDf['subtype'] == target
    ax.scatter(finalDf.loc[indicesToKeep, 'principal component 1']
               , finalDf.loc[indicesToKeep, 'principal component 2']
               , c = color
               , s = 50)
plt.ylim(top=4)  # adjust the top leaving bottom unchanged
plt.ylim(bottom=-1)  # adjust the bottom leaving top unchanged
plt.xlim(right=5)  # adjust the top leaving bottom unchanged
plt.xlim(left=-10)  # adjust the bottom leaving top unchanged
ax.legend(targets)
ax.grid()
```
