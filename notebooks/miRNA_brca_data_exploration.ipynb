### Get brca labels from train and test set

```python
brca_train = pd.read_csv("../data/TCGA_source/TCGA817-FULL_SET.csv")
brca_test = pd.read_csv("../data/TCGA_source/PAN CANCER ATLAS TCGA- FINAL TEST SET-236.csv")
```

```python
brca_train = brca_train[["TCGA_ID", "Ciriello_subtype"]]
brca_test = brca_test[["TCGA_ID", "expert_PAM50_subtypes"]]
```

```python
brca_train.head(5)
```

```python
data_train_common = data_train_common.rename(columns={"tcga_id": "TCGA_ID"})
data_test_common= data_test_common.rename(columns={"tcga_id": "TCGA_ID"})
```

```python
mirna_train_labeled = pd.merge(data_train_common, brca_train, on="TCGA_ID")
mirna_test_labeled = pd.merge(data_test_common, brca_test, on="TCGA_ID")
```

## Normalize the data

#### Taking the labels away

```python
y_train = mirna_train_labeled.Ciriello_subtype
id_train = mirna_train_labeled.TCGA_ID
mirna_train_labeled.drop(["Ciriello_subtype", "TCGA_ID"], axis=1, inplace=True)
y_test = mirna_test_labeled.expert_PAM50_subtypes
id_test = mirna_test_labeled.TCGA_ID
mirna_test_labeled.drop(["expert_PAM50_subtypes", "TCGA_ID"], axis=1, inplace=True)
```

#### Computing TPM and log(x+1) norm

```python
mirna_train_labeled_1000000 = mirna_train_labeled*1000*1000
mirna_train_labeled_norm = (mirna_train_labeled_1000000).div(mirna_train_labeled.sum(axis=1), axis=0)

mirna_test_labeled_1000000 = mirna_test_labeled*1000*1000
mirna_test_labeled_norm = (mirna_test_labeled_1000000).div(mirna_test_labeled.sum(axis=1), axis=0)
```

```python
mirna_train_labeled_log_norm = mirna_train_labeled_norm.apply(lambda x: np.log2(x+1))
mirna_test_labeled_log_norm = mirna_test_labeled_norm.apply(lambda x: np.log2(x+1))
```

```python
mirna_train_labeled_log_norm.head(5)
```

```python
mirna_test_labeled_log_norm.head(5)
```

```python
mirna_train_final = mirna_train_labeled_log_norm
mirna_test_final = mirna_test_labeled_log_norm
```

```python
mirna_train_final_subtype = pd.concat([y_train, mirna_train_final], axis=1)
mirna_test_final_subtype = pd.concat([y_test, mirna_test_final], axis=1)
```

```python
mirna_train_final_subtype.head(5)
```

```python
print(mirna_train_final_subtype.shape)
print(mirna_test_final_subtype.shape)
```

```python
mirna_train_final_meta = pd.concat([id_train, mirna_train_final_subtype], axis=1)
mirna_test_final_meta = pd.concat([id_test, mirna_test_final_subtype], axis=1)
```
