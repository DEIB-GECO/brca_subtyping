## Notebook to prepare the data to train different models only with labeled data, by labelling the non-brca samples with the specific cancer type

```python
import pandas as pd
import numpy as np
```

```python
# BRCA data is fine and ready, we just have to load it

X_brca_train = pd.read_pickle("../data/tcga_brca_raw_19036_row_log_norm_train.pkl")
```

```python
# The rest of the data will require to match the samples with the respective labels

X_tcga_no_brca = pd.read_pickle("../data/tcga_raw_no_labelled_brca_row_normalized.pkl")
X_tcga_no_brca_log_norm = pd.read_pickle("../data/tcga_raw_no_labelled_brca_log_row_normalized.pkl")
```

```python
X_tcga_no_brca.head(5)
```

```python
X_tcga_no_brca_log_norm.head(5)
```

```python
labels = X_tcga_no_brca["cancer_type"]
```

```python
X_tcga_no_brca_log_norm["subtype"] = labels
```

```python
X_brca_train.drop(["tcga_id", "sample_id", "cancer_type"], axis=1, inplace=True)
```

```python
X_brca_train.columns
```

```python
X_brca_train = X_brca_train.rename(index=str, columns={"Ciriello_subtype": "subtype"})
```

```python
X_brca_train.head(5)
```

```python
X_tcga_labeled = pd.concat([X_brca_train, X_tcga_no_brca_log_norm], sort=True)
```

```python
X_tcga_labeled.to_pickle("../data/tcga_raw_all_labelled_19036_row_log_norm_train.pkl")
```

```python
mask = X_tcga_labeled["subtype"]=="Breast Invasive Carcinoma"
```

```python
X_tcga_labeled_no_random_brca = X_tcga_labeled[~mask]
```

```python
X_tcga_labeled_no_random_brca.to_pickle("../data/tcga_raw_all_labelled_no_random_brca_19036_row_log_norm_train.pkl")
```
