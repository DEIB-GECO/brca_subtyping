# Merge CNA and RNA training data for the models

```python
import pandas as pd
import numpy as np
```

## Start with BRCA data

```python
cna_train = pd.read_pickle("../data/cna_brca_train_0.8_threshold_0.6_chrX.pkl")
cna_test = pd.read_pickle("../data/cna_brca_test_0.8_treshold_0.6_chrX.pkl")
```

```python
rna_train = pd.read_pickle("../data/tcga_brca_raw_19036_row_log_norm_train.pkl")
rna_test = pd.read_pickle("../data/tcga_brca_raw_19036_row_log_norm_test.pkl")
```

### BRCA Train

```python
rna_train.head()
```

```python
rna_train.drop(["sample_id", "Ciriello_subtype", "cancer_type"], axis=1, inplace=True)
```

```python
cna_train.head()
```

```python
merged = pd.merge(cna_train,rna_train, on="tcga_id")
```

```python
merged.dtypes[merged.dtypes!='float64']
```

```python
merged.head()
```

```python
merged.to_pickle("../data/hybrids/tcga_brca_cna_rna_meta_train.pkl")
```
