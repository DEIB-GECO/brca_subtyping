```python
rna_train = pd.read_pickle("../data/tcga_brca_raw_19036_row_log_norm_train.pkl")
rna_test = pd.read_pickle("../data/tcga_brca_raw_19036_row_log_norm_test.pkl")
```

```python
mirna_brca_train = pd.read_csv("../data/miRNA_filtered_norm_scaled_meta_train.csv")
mirna_brca_test = pd.read_csv("../data/miRNA_filtered_norm_scaled_meta_test.csv")
```

### BRCA Train
